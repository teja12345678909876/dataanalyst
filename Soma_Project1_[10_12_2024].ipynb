2. Data Cleaning

1. If there are duplicates, remove them.

2. Handle missing values appropriately (drop or impute them).

3. Rename the columns to more meaningful names, if necessary

```python
# prompt: Data Cleaning
# 1. If there are duplicates, remove them.
# 2. Handle missing values appropriately (drop or impute them).
# 3. Rename the columns to more meaningful names, if necessary.
# Deliverables:
# · Cleaned dataset.
# · Explanation of how you handled missing values and duplicates.

import opendatasets as od
import pandas as pd

# Install opendatasets if not already installed
!pip install opendatasets

# Load the dataset (replace with your actual file path if needed)
data = pd.read_excel('/content/Retail_Sales_Dataset.xlsx')

# Data Cleaning

# 1. Handling Duplicates
print("Number of duplicate rows before removal:", data.duplicated().sum())
data.drop_duplicates(inplace=True)
print("Number of duplicate rows after removal:", data.duplicated().sum())

# 2. Handling Missing Values
# Inspect missing values
print("\nMissing values before handling:\n", data.isnull().sum())

# Strategy: Fill numerical columns with the mean and categorical columns with the mode.
# Identify numerical and categorical columns
numerical_cols = data.select_dtypes(include=['number']).columns
categorical_cols = data.select_dtypes(exclude=['number']).columns

# Fill missing numerical values with the mean
for col in numerical_cols:
  if data[col].isnull().any():
    data[col].fillna(data[col].mean(), inplace=True)

# Fill missing categorical values with the mode
for col in categorical_cols:
  if data[col].isnull().any():
    data[col].fillna(data[col].mode()[0], inplace=True)

print("\nMissing values after handling:\n", data.isnull().sum())

# 3. Renaming Columns (example)
# Check for columns that need renaming and rename them appropriately.
# Example:  If a column named 'Sales_Value' needs renaming to 'SalesAmount'
# data.rename(columns={'Sales_Value': 'SalesAmount'}, inplace=True)


# Display the cleaned dataset
print("\nCleaned Dataset:")
print(data.head(10))

# Explanation of handling missing values and duplicates:
print("\nExplanation of Data Cleaning:")
print("Duplicate rows were removed using the drop_duplicates() method.")
print("Missing values were handled differently based on the data type of the column.")
print("Numerical columns' missing values were imputed using the mean of that column.")
print("Categorical columns' missing values were imputed using the mode of that column.")
```

```python
# prompt: import pandas as pd
# # Load the dataset
# df = pd.read_excel("Retail_Sales_Dataset.xlsx")
# # Convert the 'Order_Date' column to datetime format
# df['Order_Date'] = pd.to_datetime(df['Order_Date'])
# # Create new columns
# df['Day_of_Week'] = df['Order_Date'].dt.day_name()
# df['Day_of_Month'] = df['Order_Date'].dt.day
# df['Month'] = df['Order_Date'].dt.month_name()
# df['Year'] = df['Order_Date'].dt.year
# # Function to determine the season
# def get_season(date):
#     if date.month in [12, 1, 2]:
#         return 'Winter'
#     elif date.month in [3, 4, 5]:
#         return 'Spring'
#     elif date.month in [6, 7, 8]:
#         return 'Summer'
#     else:
#         return 'Fall'
# # Apply the season determination
# df['Season'] = df['Order_Date'].apply(get_season)
# # Set 'Order_Date' as the DataFrame index
# df.set_index('Order_Date', inplace=True)
# # Save the updated dataset back to a file (optional)
# df.to_excel("Updated_Retail_Sales_Dataset.xlsx")
# # Display the first few rows of the updated DataFrame
# print(df.head()) use this code 1. Convert the Order_Date column to a datetime format.
# 2. Create new columns:
# o Day of the week.
# o Day of the month.
# o Month.
# o Year.
# o Season (Winter, Spring, Summer, Fall).
# 3. Set the Order_Date column as the DataFrame index.

# Assuming 'data' DataFrame is already loaded and cleaned from the previous code

# Convert 'Order_Date' to datetime if not already done
data['Order_Date'] = pd.to_datetime(data['Order_Date'])

# Create new columns
data['Day_of_Week'] = data['Order_Date'].dt.day_name()
data['Day_of_Month'] = data['Order_Date'].dt.day
data['Month'] = data['Order_Date'].dt.month_name()
data['Year'] = data['Order_Date'].dt.year

# Function to determine the season
def get_season(date):
    if date.month in [12, 1, 2]:
        return 'Winter'
    elif date.month in [3, 4, 5]:
        return 'Spring'
    elif date.month in [6, 7, 8]:
        return 'Summer'
    else:
        return 'Fall'

# Apply the season determination
data['Season'] = data['Order_Date'].apply(get_season)

# Set 'Order_Date' as the DataFrame index
data.set_index('Order_Date', inplace=True)

# Display the first few rows of the updated DataFrame
print(data.head())
```
