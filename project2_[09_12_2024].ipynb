# ***Create a new column for the age of the car using the registration field.***

```python
import datetime

# Assuming 'data' DataFrame is already loaded and 'registration' is datetime
def calculate_car_age(registration_date):
    today = datetime.date.today()
    # Convert 'today' to Timestamp for comparison or use registration_date.date()
    if pd.isnull(registration_date) or registration_date > pd.Timestamp(today):  # Handle missing or invalid dates
        return None
    age = today.year - registration_date.year
    return age

data['car_age'] = data['registration'].apply(calculate_car_age)
data
```

```python
# prompt: Generate a price_per_km column to assess cost efficiency.

# Calculate price_per_km
data['price_per_km'] = data['price'] / data['km']

# Display the updated DataFrame with the new column
data
```

# **Create categories for cars based on price (e.g., low, mid, high).**

```python
# prompt: Create categories for cars based on price (e.g., low, mid, high).

# Assuming 'data' DataFrame is already loaded and 'price' column exists
import pandas as pd

def categorize_price(price):
    if price <= 10000:
        return 'low'
    elif price <= 20000:
        return 'mid'
    else:
        return 'high'

data['price_category'] = data['price'].apply(categorize_price)

# Display the updated DataFrame
data
```

# **Group the vehicles based on attributes like price, km, and age.**

```python
# prompt: Group the vehicles based on attributes like price, km, and age.

# Assuming 'data' DataFrame is already loaded and preprocessed as in the previous code
# ... (previous code to load and preprocess data)

# Define price and km ranges for grouping
price_bins = [0, 10000, 20000, float('inf')]
km_bins = [0, 50000, 100000, float('inf')]
age_bins = [0, 5, 10, float('inf')]

# Create labels for the bins
price_labels = ['Low', 'Medium', 'High']
km_labels = ['Low', 'Medium', 'High']
age_labels = ['New', 'Used', 'Old']

# Use cut function to create categorical columns
data['price_group'] = pd.cut(data['price'], bins=price_bins, labels=price_labels, right=False)
data['km_group'] = pd.cut(data['km'], bins=km_bins, labels=km_labels, right=False)
data['age_group'] = pd.cut(data['car_age'], bins=age_bins, labels=age_labels, right=False)

# Group the vehicles and display the groups (example)
grouped_data = data.groupby(['price_group', 'km_group', 'age_group']).size().reset_index(name='count')
grouped_data
```
